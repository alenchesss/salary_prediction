```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import pickle
import os

df = pd.read_csv('../data/processed/merged_jobs.csv')
print("Loaded:", df.shape)
df.head()
```

```python
# Remove outliers
q1 = df['salary_in_usd'].quantile(0.01)
q99 = df['salary_in_usd'].quantile(0.99)
df = df[(df['salary_in_usd'] >= q1) & (df['salary_in_usd'] <= q99)]
print("After outlier removal:", df.shape)

# Fill missing experience level
df['experience_level'] = df['experience_level'].fillna('Mid-level')
df['job_category'] = df['job_category'].fillna('Data Science and Research')
```

```python
# Encode categoricals
le_title = LabelEncoder()
le_country = LabelEncoder()
le_level = LabelEncoder()
le_category = LabelEncoder()

df['title_enc'] = le_title.fit_transform(df['job_title'])
df['country_enc'] = le_country.fit_transform(df['company_location'])
df['level_enc'] = le_level.fit_transform(df['experience_level'])
df['category_enc'] = le_category.fit_transform(df['job_category'])

print("Encoding done!")
print("Experience levels:", list(le_level.classes_))
```

```python
# Save
df.to_csv('../data/processed/clean_jobs.csv', index=False)

os.makedirs('../model', exist_ok=True)
with open('../model/label_encoders.pkl', 'wb') as f:
    pickle.dump({
        'title': le_title,
        'country': le_country,
        'level': le_level,
        'category': le_category
    }, f)

print("Saved! Shape:", df.shape)
```
