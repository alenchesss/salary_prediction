```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

df = pd.read_csv('../data/processed/clean_jobs.csv')
os.makedirs('../data/model_output', exist_ok=True)
print("Loaded:", df.shape)
```

```python
# Top 10 job titles by count
plt.figure(figsize=(10,5))
top_titles = df['job_title'].value_counts().head(10)
sns.barplot(x=top_titles.values, y=top_titles.index, palette='Blues_r')
plt.title('Top 10 Most Common Job Titles')
plt.xlabel('Number of Jobs')
plt.tight_layout()
plt.savefig('../data/model_output/top_titles.png')
plt.show()
```

```python
# Average salary by country (top 15)
plt.figure(figsize=(10,5))
avg_sal = df.groupby('company_location')['salary_in_usd'].mean().sort_values(ascending=False).head(15)
sns.barplot(x=avg_sal.values, y=avg_sal.index, palette='Oranges_r')
plt.title('Average Salary by Country (Top 15)')
plt.xlabel('Average Salary (USD)')
plt.tight_layout()
plt.savefig('../data/model_output/salary_by_country.png')
plt.show()
```

```python
# Salary distribution
plt.figure(figsize=(10,4))
sns.histplot(df['salary_in_usd'], bins=40, kde=True, color='steelblue')
plt.title('Salary Distribution (USD)')
plt.xlabel('Salary (USD)')
plt.tight_layout()
plt.savefig('../data/model_output/salary_distribution.png')
plt.show()
```

```python
# Salary by experience level
plt.figure(figsize=(8,4))
order = ['Entry-level', 'Mid-level', 'Senior', 'Executive']
sns.boxplot(data=df, x='experience_level', y='salary_in_usd', order=order, palette='Set2')
plt.title('Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.tight_layout()
plt.savefig('../data/model_output/salary_by_level.png')
plt.show()
```

```python
# Salary trend by year
plt.figure(figsize=(8,4))
yearly = df.groupby('work_year')['salary_in_usd'].mean()
plt.plot(yearly.index, yearly.values, marker='o', color='green', linewidth=2)
plt.title('Average Salary Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Average Salary (USD)')
plt.tight_layout()
plt.savefig('../data/model_output/salary_trend.png')
plt.show()
```
